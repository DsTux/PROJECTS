** Hard: What are the top 5 most popular email providers/hosts (e.g. gmail.com, yahoo.com, etc...) **

```python
ecom["Email"]
```

```python
example = ecom["Email"].iloc[0]
example
```

```python
ecom["Email"].apply(lambda x: x.split("@")[1]).value_counts().head(5)
```

```python
ecom["Email"].apply(lambda x: x.split("@")[0]).value_counts().head(5)
```

```python
domain = {}
for i in ecom["Email"]:
    if i.split('@')[1] not in domain:
        domain[i.split('@')[1]] =1
    else:
        domain[i.split('@')[1]] +=1
pd.Series(domain).sort_values(ascending =False).head(5)
```

```python
import re
ecom["Email"].str.extract('(@\w+.\w*)')[0].value_counts().head(5)
```
