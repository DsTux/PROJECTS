```python
# At the begining we can focus on the columns that have %35 percantage and higher of missing values.
def show_nans(oto, limit):
    missing = oto.isnull().sum()*100/oto.shape[0]
    return missing.loc[lambda x : x > limit]
show_nans(oto,35)
```

```python
# we can drop the columns which have 90 percentage or high missing values
```

```python
def fill_most(oto, group_col, col_name):
    for group in list(oto[group_col].unique()):
        grp_inx = list(oto[oto[group_col] == group][col_name].index)
        oto[col_name].iloc[grp_inx] = oto[col_name].iloc[grp_inx].fillna(oto[oto[group_col] == group][col_name].mode())
        
    
```

```python
fill_most(oto, "make_model", "body_type")
```
