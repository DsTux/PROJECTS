#### dropping columns

```python
#drop columns which have lots NAN values 
drop_columns = ["kW", "Next Inspection", "Inspection new", "Emission Label", "Electricity consumption", "Last Service Date", "Other Fuel Types", "Availability", "Last Timing Belt Service Date", "Available from", "url", "null", "Previous Owners", "registration", "Emission Class", "description", "Non-smoking Vehicle", "Model Code", "Weight"]
df.drop(drop_columns, axis = "columns", inplace = True)
```

```python
df.shape
```

##### drop the /n if the column has /n (general formule)

```python
def n_sil(x):
    for i in range(0, len(x)):
        if isinstance(x[i], list):
            for a in x[i]:
                if a == "\n":
                    x[i].remove(a)
                elif "\n" in a:
                    x[i].remove(a)
                    b = a.replace("\n", "")
                    x[i].append(b)
        elif isinstance(x[i], str):
            x[i] = x[i].replace("\n", "")
```

## General formul to make numeric

```python
def coerce_m_columns_to_numeric(m, column_list):
    m[column_list] = m[column_list].apply(pd.to_numeric, errors='coerce')
```
